$ \large{\textbf{1. Stellar Profiles}}$

$\textit{MESA files and logs are available in this github.}$

```python
import matplotlib.pyplot as plt
import numpy as np
import mesa_reader as mr
```

```python
class Model:
    def __init__(self, mass, profiles=None, history=None):
        self.mass = mass
        #if history:
        #    self.history = mr.MesaData(history)
        self.profiles = []
        if profiles:
            for p in profiles:
                self.profiles.append(mr.MesaData(p))
```

```python
zams = Model(8, profiles=["./M8_at_ZAMS/mesa_sim/LOGS/profile1.data"], history=["./M8_at_ZAMS/mesa_sim/LOGS/history.data"])
hb = Model(8, profiles=["./M8_at_HB/mesa_sim/LOGS/profile1.data"], history=["./M8_at_ZAMS/mesa_sim/LOGS/history.data"])
```

```python
# columns in MesaData
zams.profiles[0].logR
hb.profiles[0].logR

scaled_zams = zams.profiles[0].logR - zams.profiles[0].logR[0]
scaled_hb = hb.profiles[0].logR - hb.profiles[0].logR[0]
```

$ \large{\textbf{1.a}}$

```python
plt.figure(figsize=(12, 8))

# ZAMS profile
plt.plot(scaled_zams, zams.profiles[0].logRho, label='ZAMS Phase', color='#1f77b4', linewidth=3, marker='o', markersize=6, markerfacecolor='white')

# Helium burning profile
plt.plot(scaled_hb, hb.profiles[0].logRho, label='Helium Burning Phase', color='#ff7f0e', linewidth=3, linestyle='--', marker='s', markersize=6, markerfacecolor='white')

plt.xlabel(r'$\log_{10}(r/R)$', fontsize=18, fontweight='bold', labelpad=10)
plt.ylabel(r'$\log_{10}(\rho\ [g\ cm^{-3}])$', fontsize=18, fontweight='bold', labelpad=10)
plt.title('Density Profile: ZAMS vs Helium Burning Phase', fontsize=20, fontweight='bold', pad=20)

plt.grid(True, which="both", linestyle='--', linewidth=1.2, alpha=0.7)

plt.yscale('linear')
plt.xlim(-3, 0)

plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

plt.legend(loc='upper right', fontsize=14, frameon=True, fancybox=True, shadow=True, borderpad=1, framealpha=0.9)

plt.tight_layout()
plt.savefig("plots/density.png")
plt.show()
```

The plot compares the star's density profiles during the ZAMS and Helium Burning phases. In the ZAMS phase, the core is dense, with a sharp decline outward. During Helium Burning, the core density decreases significantly, indicating internal restructuring as fusion shifts from hydrogen to helium. The outer regions maintain similar densities in both phases.

$ \large{\textbf{1.b}}$

```python
plt.figure(figsize=(12, 8))

# ZAMS profile
plt.plot(scaled_zams, zams.profiles[0].logT, label='ZAMS Phase', color='#1f77b4', linewidth=3, marker='o', markersize=6, markerfacecolor='white')

# Helium burning profile
plt.plot(scaled_hb, hb.profiles[0].logT, label='Helium Burning Phase', color='#ff7f0e', linewidth=3, linestyle='--', marker='s', markersize=6, markerfacecolor='white')

plt.xlabel(r'$\log_{10}(r/R)$', fontsize=14)
plt.ylabel(r'$\log_{10}(T\ [K])$', fontsize=14)
plt.title('Temperature Profile: ZAMS vs Helium Burning Phase', fontsize=16)

plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.xlim(-3, 0)  
plt.ylim(5, 8)   

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend()

plt.tight_layout()
plt.savefig("plots/temperature.png")
plt.show()
```

The temperature profile reflects the internal processes of stellar evolution. In the ZAMS (Zero-Age Main Sequence) phase, hydrogen fusion occurs in the core, maintaining a relatively stable and high temperature throughout the star's interior. This stable temperature reflects the star's balance between gravitational collapse and nuclear fusion pressure.

In the Helium Burning phase, after the hydrogen fuel is exhausted, the core contracts, leading to helium fusion in a much denser and hotter core. However, as helium fusion is less efficient than hydrogen fusion, the outer layers cool down slightly. The significant temperature drop in the outer regions is due to the expansion and redistribution of energy, as the star adjusts to burning heavier elements.

$ \large{\textbf{1.c}}$

```python
plt.figure(figsize=(12, 8))

# ZAMS profile
plt.plot(10**scaled_zams, zams.profiles[0].mass, label='ZAMS Phase', color='#1f77b4', linewidth=3, marker='o', markersize=6, markerfacecolor='white')

# Helium burning profile
plt.plot(10**scaled_hb, hb.profiles[0].mass, label='Helium Burning Phase', color='#ff7f0e', linewidth=3, linestyle='--', marker='s', markersize=6, markerfacecolor='white')

plt.xlabel(r'$(r/R_{sun})$', fontsize=18, fontweight='bold', labelpad=10)
plt.ylabel(r'$m(r)$ (in $M_{sun}$)', fontsize=18, fontweight='bold', labelpad=10)
plt.title('Mass Profile: ZAMS vs Helium Burning Phase', fontsize=20, fontweight='bold', pad=20)

plt.grid(True, which="both", linestyle='--', linewidth=1.2, alpha=0.7)

plt.yscale('linear')
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

plt.legend(loc='lower right', fontsize=14, frameon=True, fancybox=True, shadow=True, borderpad=1, framealpha=0.9)

plt.tight_layout()
plt.savefig("plots/mass.png")
plt.show()
```

During ZAMS, the star is stable, fusing hydrogen in its core, and mass increases steadily from the core to the surface. The core is small, with most mass concentrated near the center. In the Helium Burning phase, the core contracts and becomes denser, while outer layers expand. Mass in the core increases rapidly, indicating higher density and concentration. Beyond the core, mass distribution flattens as outer layers expand and contain less mass. This shows that as the star evolves, mass becomes more concentrated in the core, while outer layers expand and contribute less to the overall mass distribution.

$ \large{\textbf{1.e}}$

```python
r_over_R_zams = 10**scaled_zams  # r/R values for ZAMS
x_mass_fraction_H_zams = zams.profiles[0].x_mass_fraction_H  # Hydrogen mass fraction (ZAMS)
y_mass_fraction_He_zams = zams.profiles[0].y_mass_fraction_He  # Helium mass fraction (ZAMS)
z_mass_fraction_metals_zams = zams.profiles[0].z_mass_fraction_metals  # Metal mass fraction (ZAMS)

r_over_R_hb = 10**scaled_hb  # r/R values for Helium burning
x_mass_fraction_H_hb = hb.profiles[0].x_mass_fraction_H  # Hydrogen mass fraction (HB)
y_mass_fraction_He_hb = hb.profiles[0].y_mass_fraction_He  # Helium mass fraction (HB)
z_mass_fraction_metals_hb = hb.profiles[0].z_mass_fraction_metals  # Metal mass fraction (HB)


fig, axs = plt.subplots(1, 3, figsize=(18, 6))  # 3 subplots in 1 column

# Subplot 1: Hydrogen Mass Fraction
axs[0].plot(r_over_R_zams, x_mass_fraction_H_zams, label='ZAMS Phase', color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=6, markerfacecolor='white')
axs[0].plot(r_over_R_hb, x_mass_fraction_H_hb, label='Helium Burning Phase', color='#1f77b4', linewidth=3, linestyle='--', marker='o', markersize=6)
axs[0].set_xlabel(r'$(r/R_{sun})$', fontsize=14)
axs[0].set_ylabel('Hydrogen (H) Mass Fraction', fontsize=14)
axs[0].legend(loc='best', fontsize=12)
axs[0].grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)

# Subplot 2: Helium Mass Fraction
axs[1].plot(r_over_R_zams, y_mass_fraction_He_zams, label='ZAMS Phase', color='#ff7f0e', linewidth=3, linestyle='-', marker='s', markersize=6, markerfacecolor='white')
axs[1].plot(r_over_R_hb, y_mass_fraction_He_hb, label='Helium Burning Phase', color='#ff7f0e', linewidth=3, linestyle='--', marker='s', markersize=6)
axs[1].set_xlabel(r'$(r/R_{sun})$', fontsize=14)
axs[1].set_ylabel('Helium (He) Mass Fraction', fontsize=14)
axs[1].legend(loc='best', fontsize=12)
axs[1].grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)

# Subplot 3: Metal Mass Fraction
axs[2].plot(r_over_R_zams, z_mass_fraction_metals_zams, label='ZAMS Phase', color='#2ca02c', linewidth=3, linestyle='-', marker='^', markersize=6, markerfacecolor='white')
axs[2].plot(r_over_R_hb, z_mass_fraction_metals_hb, label='Helium Burning Phase', color='#2ca02c', linewidth=3, linestyle='--', marker='^', markersize=6)
axs[2].set_xlabel(r'$(r/R_{sun})$', fontsize=14)
axs[2].set_ylabel('Metal (Z) Mass Fraction', fontsize=14)
axs[2].legend(loc='best', fontsize=12)
axs[2].grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)

fig.suptitle('Mass Fractions vs Normalized Radius (ZAMS & Helium Burning)', fontsize=18, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("plots/fraction.png")  

plt.show()
```
